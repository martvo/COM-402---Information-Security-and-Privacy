```python
import pandas as pd
```

```python
columns = ["name", "movie", "date", "rating"]
netflix = pd.read_csv("./database/com402-1.csv", skipinitialspace=True, quotechar='"', header=None, names=columns)
imdb = pd.read_csv("./database/imdb-1.csv", skipinitialspace=True, quotechar='"', header=None, names=columns)
imdb.head()
```

```python
netflix.head()
```

```python
# find my hash
joined_df = netflix.merge(imdb, on=["date"])
joined_df_me = joined_df[joined_df["name_y"] == "martin.vold@epfl.ch"]

my_hash = joined_df_me.groupby(["name_x"]).count().sort_values(["movie_x"], ascending=False)
my_hash = my_hash.index[0]
my_hash
```

```python
# get my hashed movies whivh matches with my_hash
my_hashed_movies = netflix[netflix["name"] == my_hash]
my_hashed_movies
```

```python
# for each of my movie hashes, get all rows with the same movie hash
# choose the cleartext movie with the highest count
my_movies = []
for movie_hash in my_hashed_movies["movie"].values:
    joined = joined_df[joined_df["movie_x"] == movie_hash]
    print(joined["movie_y"].value_counts().index[0])
    my_movies.append(joined["movie_y"].value_counts().index[0])
```

```python
f = open("ex1a.txt", "a")
for movie in my_movies:
    f.write(movie+"\n")
f.close()
```
