```python
import pandas as pd
```

```python
columns = ["name", "movie", "date", "rating"]
netflix = pd.read_csv("./database/com402-2.csv", skipinitialspace=True, quotechar='"', header=None, names=columns)
imdb = pd.read_csv("./database/imdb-2.csv", skipinitialspace=True, quotechar='"', header=None, names=columns)
```

```python
netflix.head()
```

```python
# Get frequencies for each movie in the imdb and netflix table, and sort them
imdb_count = imdb.groupby(["movie"]).count().sort_values(["name"], ascending=False)
netflix_count = netflix.groupby(["movie"]).count().sort_values(["name"], ascending=False)

imdb_count
```

```python
print(len(imdb_count), len(netflix_count))
```

```python
# Create a dictionary maping from each movie hash to the plaintext of the movie
movie_hash_to_movie = {}
for movie, movie_hash in zip(imdb_count.index, netflix_count.index):
    movie_hash_to_movie[movie_hash] = movie
```

```python
# Map every user to a hash based on the movies we have, if the imdb movies is a subset of the netflix movies we have found
# the user
user_to_hash = {}

for user in imdb.name.unique():
    user_movies = set(imdb[imdb.name == user].movie.values)
        
    for hash_user in netflix.name.unique():
        hash_movies = set([movie_hash_to_movie[m] for m in netflix[netflix.name == hash_user].movie.values])
        if len(user_movies - hash_movies) == 0:
            user_to_hash[user] = hash_user
            
user_to_hash
```

```python
my_hashed_movies = netflix[netflix.name == user_to_hash["martin.vold@epfl.ch"]].movie

f = open("ex1b.txt", "a")
for movie in my_hashed_movies:
    f.write(movie_hash_to_movie[movie]+"\n")
f.close()
```
